```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on= "Country Code", how='left')
```

```python
final_df
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names  = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# Pie Chart Top 3 Countries that uses Zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct="%1.1f%%")
```

obeservations: Zomato Maximum records or transaction are from India after then from united states then united kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
ratings.head()
```

```python
sns.barplot(x = "Aggregate rating", y="Rating Count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings,hue='Rating color', palette=['grey','red','orange','yellow', 'green' , 'green'])
```

Observation 
1. Not Rated Count is very High
2. Maximum Number of ratings are between 2.5 to 3.4

```python
## Count Plot
sns.countplot(x="Rating color", data=ratings, palette=['grey','red','orange','yellow', 'green' , 'green'])
```
